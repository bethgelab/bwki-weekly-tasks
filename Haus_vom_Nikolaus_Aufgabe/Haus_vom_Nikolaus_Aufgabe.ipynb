## Jetzt wird es mathematisch
Es wurde mathematisch bewiesen, dass es tatsächlich nur von den Punkten 0 und 1 möglich ist, das Haus in einem Pinselstrich zu zeichnen. Wenn man bei Punkt 0 anfängt und ein korrektes Haus zeichnet, dann kommt man immer bei Punkt 1 heraus. Das ist auch so im Video zu sehen. Und wenn man bei Punkt 1 anfängt, dann kommt man immer bei Punkt 0 heraus. Demnach sind die Lösungen, die bei Punkt 1 beginnen nicht verschieden, sondern nur gespiegelt zu Lösungen, die in Punkt 0 anfangen. Im Video sind 44 Möglichkeiten (mit Spiegelung wären das 88 Möglichkeiten) zu sehen, aber sind das wirklich alle? Dieser Frage wollen wir in dieser Aufgabe nachgehen.

Dazu müssen wir systematisch alle korrekten Möglichkeiten, das Haus zu zeichnen aufzählen. Wir betrachten das Haus vom Nikolaus als einen sogenannten Graphen. Ein Graph in der Mathematik hat sogenannte Ecken (in unserem Bild sind das die roten Punkte) und Kanten, die die Ecken verbinden (in unserem Fall die Möglichkeiten für Linien des Hauses).

Man kann für einen Graphen auch aufschreiben, welche Ecken mit welchen anderen Ecken über Kanten verbunden sind:
 - (0-1), (0-2), (0-3)
 - (1-0), (1-2), (1-3)
 - (2-0), (2-1), (2-3), (2-4)
 - (3-0), (3-1), (3-2), (3-4)
 - (4-2), (4-3)
 
Man kann die Kanten auch zu sogenannten Adjazenzlisten für die Ecken zusammenfassen:
 - 0 -> (1, 2, 3)
 - 1 -> (0, 2, 3)
 - 2 -> (0, 1, 3, 4)

Von Ecke 0 kann man also nach Ecke 1, 2, oder 3 eine Kante zeichnen. Von Ecke 1 kann man wieder zurück zu Ecke 0 oder auch zu Ecke 2 und 3. Und von dort aus kann man wieder neue Ecken erreichen und sich so durch den Graphen 'hangeln'.

Mit dieser Information kann nun ein Suchalgorithmus alle Möglichkeiten aufzählen, das Haus zu zeichnen. Um eine korrekte Möglichkeit zu finden, läuft man bei Ecke 0 los, muss im Verlauf alle Kanten zeichnen und zuletzt bei Ecke 1 ankommen. Eine Kante darf dabei nie doppelt verwendet werden. Zudem darf eine bereits gefundene korrekte Möglichkeit das Haus zu zeichen nicht noch einmal gefunden und somit doppelt gezählt werden.

Ein solches Suchproblem in Graphen lässt sich mit der Brute-Force-Methode oder eleganter mit einem rekursiven Algorithmus lösen. Dazu folgen später weitere Erklärungen.

Außerdem müssen alle gefundenen korrekten Möglichkeiten zum Zeichnen des Hauses abgespeichert werden, um Duplikate auszuschließen und die Anzahl zum Schluss zu zählen. Ein einfacher Weg, eine Möglichkeit abzuspeichern ist eine Liste der Ecken, die nacheinander besucht werden. So ist die Liste [0, 2, 1, 0, 3, 4, 2, 3, 1] eine valide Möglichkeit, das Haus zu zeichen. Das kannst du mit der Funktion im nächsten Codeblock überprüfen. Diese erhält eine Liste der Ecken und zeichnet ein Haus vom Nikolaus daraus.

```python
# das Haus vom Nikolaus wird bei uns mit Matplotlib gezeichnet
import matplotlib
import matplotlib.pyplot as plt

# zunächst wird eine Methode vorgegeben, die eine Visualisierung des Hauses anzeigt
# die Methode bekommt dafür eine Liste der nacheinander besuchten Ecken übergeben
def visualisiere_haus_vom_nikolaus(ecken_liste):
    # das sind die Koordinaten der Ecken
    ecken = [(0, 0), (1, 0), (1, 1), (0, 1), (0.5, 1.5)]

    # in dieser Funktion werden die schwarzen Eckpunkte gezeichnet
    def male_alle_eckpunkte():
        for e in ecken:
            plt.plot(e[0], e[1], 'ko')
    
    # in dieser Funktion werden zwei Eckpunkte durch eine farbige Linie verbunden
    def verbinde_zwei_ecken(e1, e2, index):
        if index > 7:
            print("Fehler: so viele Kanten darf man gar nicht zeichnen!")
            return
        # Matplotlib zeichnen linien nicht mit den Koordinaten, wie sie logisch wären
        # deshalb muss man die Koordinaten etwas tauschen
        a = ecken[e1][0]
        b = ecken[e1][1]
        c = ecken[e2][0]
        d = ecken[e2][1]
        # zeichne die Ecke in einer Farbe, so dass man den Pinselstrich nachvollziehen kann 
        farbe = matplotlib.cm.get_cmap('viridis')(index / 6)
        plt.plot([a, c], [b, d], color = farbe, label = str(index) + '. Linie')

    # male alle Pinselstriche und die Eckpunkte
    for i in range(len(ecken_liste) - 1):
        verbinde_zwei_ecken(ecken_liste[i], ecken_liste[i + 1], i)
    male_alle_eckpunkte()
    
    # zeige den Plot an
    plt.legend()
    plt.axis('equal')
    plt.xticks([])
    plt.yticks([])
    plt.show()

# teste die Funktion mit dem Beispiel von oben
visualisiere_haus_vom_nikolaus([0, 2, 1, 0, 3, 4, 2, 3, 1])
```
